```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
v=pd.read_csv(r'/content/nykaa_campaign_data.csv')
```

```python
v.drop_duplicates(inplace=True)
```

```python
v['CTR'] = (v['Clicks'] / v['Impressions']) * 100
```

```python
v['Conversion_Rate'] = (v['Conversions'] / v['Clicks']) * 100
```

```python
v['CPC'] = v['Acquisition_Cost'] / v['Clicks']
```

```python
v['Revenue_Per_Conversion'] = (
    v['Revenue'] / v['Conversions']
)
```

```python
campaign_roi = v.groupby(
    'Campaign_Type'
)['ROI'].mean()

plt.figure(figsize=(8,8))

plt.pie(
    campaign_roi,
    labels=campaign_roi.index,
    autopct='%1.1f%%',
    startangle=90,
    explode=[0,0,0,0,0.1],
    colors=['lavender','pink','purple','skyblue','yellow'],shadow=True,
   )

plt.title(
    'Average ROI Distribution by Campaign Type',
    fontsize=14
)
plt.legend(loc='lower left')

plt.show()
```

```python
print(campaign_roi.index.tolist())
```

```python
import matplotlib.ticker as ticker

top_channels = v['Channel_Used'].value_counts().head(10).index

filtered_df = v[
    v['Channel_Used'].isin(top_channels)
]

channel_revenue = filtered_df.groupby(
    'Channel_Used'
)['Revenue'].mean().sort_values(
    ascending=False
).reset_index()

plt.figure(figsize=(12,6))

sns.barplot(
    x='Channel_Used',
    y='Revenue',
    data=channel_revenue
)

plt.title(
    'Top 10 Marketing Channels by Average Revenue',
    fontsize=14
)

plt.xlabel('Marketing Channel')
plt.ylabel('Average Revenue')

# Rotate labels
plt.xticks(rotation=30)

# Format revenue values
plt.gca().yaxis.set_major_formatter(
    ticker.FuncFormatter(
        lambda x, pos: f'{x/1000:.0f}K'
    )
)

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(
    x='Target_Audience',
    y='Conversions',
    data=v
)

plt.xticks(rotation=45)

plt.title('Audience Conversion Analysis')

plt.show()
```

```python
plt.figure(figsize=(12,8))

numeric_df = v.select_dtypes(include=np.number)

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='viridis'
)

plt.title('Correlation Heatmap')

plt.show()
```
